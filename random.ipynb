```python
# ...existing code...

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Generate 100 random x values
np.random.seed(0)
x = np.random.rand(100) * 100  # x in range [0, 100]
# Linear function: y = 3x + 7 + noise
noise = np.random.randn(100) * 10
y = 3 * x + 7 + noise

# --- Using numpy (manual calculation) ---
mean_x = np.mean(x)
mean_y = np.mean(y)
m = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x) ** 2)
c = mean_y - m * mean_x
y_hat_np = m * x + c

# --- Using scipy ---
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
y_hat_scipy = slope * x + intercept

print("Manual (numpy) Slope:", m)
print("Manual (numpy) Intercept:", c)
print("Scipy Slope:", slope)
print("Scipy Intercept:", intercept)
print(f"R-value: {r_value}")
print(f"P-value: {p_value}")
print(f"Standard error: {std_err}")

# Plot both regression lines
plt.scatter(x, y, color='lightblue', label='Data points')
plt.plot(x, y_hat_np, color='green', label='Numpy Regression')
plt.plot(x, y_hat_scipy, color='red', linestyle='--', label='Scipy Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of Linear Regression (Numpy vs Scipy)')
plt.legend()
plt.show()

# Try modifying y to see effect on R, P, std_err
# For example, add more noise:
y_noisy = 3 * x + 7 + np.random.randn(100) * 50
slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(x, y_noisy)
print("\nWith more noise:")
print(f"R-value: {r_value2}")
print(f"P-value: {p_value2}")
print(f"Standard error: {std_err2}")

# ...existing code...
```
