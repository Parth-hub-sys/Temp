```python
# Train a logistic regression model from SKlearn library for classification of data of the
# Iris dataset. Also perform the steps (b)-(e) from the above practical(5). Also display
# various metrics â€“ Accuracy, Precision_score, recall_score, f1_score.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, ConfusionMatrixDisplay

# a. Load Iris dataset
iris = load_iris()
X = iris.data
Y = iris.target

# b. Generate training and testing datasets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# c. Train the logistic regression model
model = LogisticRegression(max_iter=200)
model.fit(X_train, Y_train)

# d. Predict Y for test dataset
Y_pred = model.predict(X_test)

# e. Draw the plot (confusion matrix)
ConfusionMatrixDisplay.from_estimator(model, X_test, Y_test)
plt.title('Confusion Matrix - Iris Logistic Regression')
plt.show()

# Display metrics
print(f"Accuracy: {accuracy_score(Y_test, Y_pred):.2f}")
print(f"Precision: {precision_score(Y_test, Y_pred, average='weighted'):.2f}")
print(f"Recall: {recall_score(Y_test, Y_pred, average='weighted'):.2f}")
print(f"F1 Score: {f1_score(Y_test, Y_pred, average='weighted'):.2f}")
```
