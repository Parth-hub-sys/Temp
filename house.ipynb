```python
# Build a linear regression model that predicts house prices based on square footage.
```

```python
# ...existing code...

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Generate synthetic data: square footage (x) and house price (y)
np.random.seed(42)
square_footage = np.random.randint(500, 4000, 100)  # 100 houses, 500-4000 sqft
# Assume price = 150 * sqft + 50,000 + noise
noise = np.random.randn(100) * 20000
price = 150 * square_footage + 50000 + noise

# Build linear regression model using scipy
slope, intercept, r_value, p_value, std_err = stats.linregress(square_footage, price)
predicted_price = slope * square_footage + intercept

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-value: {r_value}")
print(f"P-value: {p_value}")
print(f"Standard error: {std_err}")

# Plotting
plt.scatter(square_footage, price, color='blue', label='Actual data')
plt.plot(square_footage, predicted_price, color='red', label='Regression line')
plt.xlabel('Square Footage')
plt.ylabel('House Price')
plt.title('Linear Regression: House Price vs Square Footage')
plt.legend()
plt.show()

# ...existing code...
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
np.random.seed(60)
sf = np.random.randint(200,1000,125)

noise = np.random.randn(125)*5000
price = 500*sf + 35000 + noise
slope,intercept,r_value,p_value,std_err = stats.linregress(sf,price)

predicted_price = slope*sf + intercept
```

```python
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-value: {r_value}")
print(f"P-value: {p_value}")
print(f"Standard error: {std_err}")
```

```python
plt.scatter(sf,price,color='blue',label='Actual data')
plt.plot(sf,predicted_price,color='red',label='Regression line')
plt.xlabel('Square Footage')
plt.ylabel('House Price')
plt.title('Linear Regression: House Price vs Square Footage')
plt.legend()
plt.show()
```
