```python
# ...existing code...
# Train a multiple linear regression model from SKlearn library for the California
# housing dataset. Also perform the steps (b)-(e) from the above practical(5).

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load California housing dataset
data = fetch_california_housing()
X = data.data  # Features (multiple)
Y = data.target  # Target (median house value)

# b. Generate training and testing datasets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# c. Train the model
model = LinearRegression()
model.fit(X_train, Y_train)

# d. Predict Y for test dataset
Y_pred = model.predict(X_test)

# e. Draw the plot (actual vs predicted)
plt.scatter(Y_test, Y_pred, color='purple', alpha=0.5)
plt.xlabel('Actual Median House Value')
plt.ylabel('Predicted Median House Value')
plt.title('Multiple Linear Regression: California Housing')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red', lw=2, label='Ideal fit')
plt.legend()
plt.show()

# ...existing code...
```
