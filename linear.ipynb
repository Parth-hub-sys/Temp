```python
# For the given input values of x and y, build a linear regression model using scipy
# library and plot it.
```

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([3,1,8,5,2])
```

```python
slope,intercept,r_value,p_value,std_err = stats.linregress(x,y)
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
```

```python
y_hat = slope * x + intercept
```

```python
plt.scatter(x,y,color='blue', label='Data points')
plt.plot(x,y_hat,color='red', label='Regression line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear Regression using scipy')
plt.legend()
plt.show()
```
