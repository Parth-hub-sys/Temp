```python
import matplotlib.pyplot as plt
import numpy as np

# First plot: two lines
plt.figure(figsize=(6, 10))

plt.subplot(2, 1, 1)
plt.plot([10, 50], [30, 30], label='line 1')  # Horizontal line at y=30
plt.plot([30, 30], [10, 50], label='line 2')  # Vertical line at x=30
plt.xlim(10, 50)
plt.ylim(10, 50)
plt.legend()
plt.show()

```

```python
# Second plot: colored horizontal lines in a grid
plt.subplot(2, 1, 2)
colors = ['red', 'green', 'yellow', 'blue']
for i in range(4):
    for j in range(10):
        alpha = 1 - j * 0.1
        plt.hlines(y=8 - 2*i, xmin=1 + i*2, xmax=2 + i*2, colors=colors[i], linewidth=5, alpha=alpha)
plt.xlim(1, 9)
plt.ylim(-8, 8)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(6)
y1 = np.array([3, 7, 12, 18, 25, 33])
y2 = np.array([2, 5, 9, 14, 10, 10])
y3 = np.array([2, 3, 4, 5, 6, 7])
y4 = np.array([2, 4, 7, 11, 14, 17])

fig, axs = plt.subplots(1, 2, figsize=(8, 5))

# Left plot
axs[0].bar(x, y1, color='tab:blue')
axs[0].bar(x, y2, color='tab:orange')

# Right plot
axs[1].bar(x, y3, color='red')
axs[1].bar(x, y4, color='yellowgreen')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Stacked Bar Plot ---
labels = ['A', 'B', 'C', 'D']
round1 = [20, 35, 30, 35]
round2 = [15, 32, 34, 20]
round3 = [10, 25, 10, 15]
round4 = [8, 20, 15, 12]

width = 0.5
x = np.arange(len(labels))

fig, axs = plt.subplots(2, 1, figsize=(7, 8))

# Stacked bars
axs[0].bar(x, round1, width, label='Round 1', color='red')
axs[0].bar(x, round2, width, bottom=round1, label='Round 2', color='blue')
axs[0].bar(x, round3, width, bottom=np.array(round1)+np.array(round2), label='Round 3', color='yellow')
axs[0].bar(x, round4, width, bottom=np.array(round1)+np.array(round2)+np.array(round3), label='Round 4', color='green')

axs[0].set_ylabel('Score')
axs[0].set_title('Scores by Teams in 4 Rounds')
axs[0].set_xticks(x)
axs[0].set_xticklabels(labels)
axs[0].legend()


# --- Pie Chart ---
sizes = [30, 25, 20, 25]
days = ['Sunday', 'Saturday', 'Friday', 'Thursday']
colors = ['black', 'pink', 'blue', 'green']

axs[1].pie(sizes, labels=days, colors=colors, startangle=90, autopct=None)
axs[1].set_title('Restaurant Sales analysis 2024', color='red', fontsize=18, pad=30)

plt.tight_layout()
plt.show()
```
