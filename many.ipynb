```python
# # ...existing code...
# 11. Build a linear regression model using SKlearn library for the following :
# a. Take input values in terms of lists (X,Y).
# b. Generate training dataset and testing dataset from the input lists using
# train_test_split() from SKlearn.
# c. Train the model using training dataset.
# d. Predict Y for test dataset.
# e. Draw the plot.
# f. Repeat step (b)-(e) for random pairs of (X,Y).

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# a. Input values as lists
X = [500, 800, 1000, 1200, 1500, 1800, 2000, 2200, 2500, 3000]
Y = [85000, 120000, 145000, 170000, 210000, 250000, 275000, 295000, 330000, 400000]

# b. Generate training and testing datasets
X_np = np.array(X).reshape(-1, 1)
Y_np = np.array(Y)
X_train, X_test, Y_train, Y_test = train_test_split(X_np, Y_np, test_size=0.3, random_state=42)

# c. Train the model
model = LinearRegression()
model.fit(X_train, Y_train)

# d. Predict Y for test dataset
Y_pred = model.predict(X_test)

# e. Draw the plot
plt.scatter(X_train, Y_train, color='blue', label='Train data')
plt.scatter(X_test, Y_test, color='green', label='Test data')
plt.plot(X_test, Y_pred, color='red', label='Predicted line')
plt.xlabel('Square Footage')
plt.ylabel('House Price')
plt.title('SKlearn Linear Regression: House Price vs Square Footage')
plt.legend()
plt.show()

# f. Repeat for random pairs of (X, Y)
np.random.seed(1)
X_rand = np.random.randint(500, 4000, 100).reshape(-1, 1)
Y_rand = 150 * X_rand.flatten() + 50000 + np.random.randn(100) * 20000

X_train2, X_test2, Y_train2, Y_test2 = train_test_split(X_rand, Y_rand, test_size=0.3, random_state=42)
model2 = LinearRegression()
model2.fit(X_train2, Y_train2)
Y_pred2 = model2.predict(X_test2)

plt.scatter(X_train2, Y_train2, color='blue', label='Train data')
plt.scatter(X_test2, Y_test2, color='green', label='Test data')
plt.plot(X_test2, Y_pred2, color='red', label='Predicted line')
plt.xlabel('Square Footage')
plt.ylabel('House Price')
plt.title('SKlearn Linear Regression: Random Data')
plt.legend()
plt.show()

# ...existing code...
```
