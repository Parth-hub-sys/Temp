```python
import numpy as np
```

```python
# For the given input values of x and y, compute coefficient values to build linear
# regression model and plot it. Use this model to predict the value of y for the new
# value of x. Also compute MSE and RSS. (Note: Use only numpy library)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])
```

```python
n = len(x)
mean_x = np.mean(x)
mean_y = np.mean(y)

numerator = np.sum((x- mean_x)* (y-mean_y))
denominator = np.sum((x-mean_x)**2)

m = numerator / denominator
c = mean_y - (m * mean_x)
```

```python
x_new = 6
y_pred = m*x_new + c
print(f"Predicted value of y for x = {x_new} is: {y_pred}")
```

```python
y_hat = m*x +c
```

```python
mse = np.mean((y-y_hat)**2)
print(f"MSE : {mse}")
```

```python
rss = np.sum((y-y_hat)**2)
print(f"RSS : {rss}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, y_hat, color='red', label='Regression line')
plt.scatter(x_new, y_pred, color='green', label='Predicted point')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
